```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('data/HousingData.csv.xls')
df.head()
```

```python
df.isnull().sum()
```

```python
imputer = SimpleImputer(strategy='mean')
df['CRIM'] = imputer.fit_transform(df['CRIM'].to_frame())
df['ZN'] = imputer.fit_transform(df['ZN'].to_frame())
df['INDUS'] = imputer.fit_transform(df['INDUS'].to_frame())
df['CHAS'] = imputer.fit_transform(df['CHAS'].to_frame())
df['AGE'] = imputer.fit_transform(df['AGE'].to_frame())
df['LSTAT'] = imputer.fit_transform(df['LSTAT'].to_frame())
```

```python
cols = ['LSTAT', 'INDUS', 'NOX', 'RM', 'MEDV']
sns.pairplot(df, size=2.5)
plt.tight_layout()
plt.show()
```

RM and MEDV have linear relationship.

```python
cm = np.corrcoef(df.values.T)
sns.set(font_scale=1.0)
sns.heatmap(cm,
            cbar=True,
            annot=True,
            square=True,
            fmt='.1f',
            annot_kws={'size': 10},
            yticklabels=df.columns,
            xticklabels=df.columns)
plt.show()
```

- The target value - MEDV - has the largets correlation with RM and LDSTAT.
